
# üß† ¬øQu√© es la programaci√≥n?

La **programaci√≥n** es el arte de indicarle a una computadora lo que tiene que hacer mediante un conjunto de instrucciones. Pod√©s imaginarlo como una **receta de cocina** üßë‚Äçüç≥: cada paso debe estar claramente definido para que la m√°quina lo entienda y lo ejecute correctamente.

Estas instrucciones se escriben en un **lenguaje especial** que llamamos **Lenguaje de Programaci√≥n**, que funciona como el **idioma de las m√°quinas** üñ•Ô∏è.

---

## üíª ¬øQu√© es un lenguaje de programaci√≥n?

Un **lenguaje de programaci√≥n** es el conjunto de instrucciones que permite a los humanos comunicarse con las computadoras. A trav√©s de √©l, podemos expresar **algoritmos** e **instrucciones** que la m√°quina puede interpretar y ejecutar.

- üß© **Sintaxis**: son las reglas y la estructura del lenguaje (c√≥mo se escribe).
- üß† **Sem√°ntica**: es el significado de esas instrucciones (qu√© hacen).

En resumen, es el puente entre nuestras ideas y la ejecuci√≥n autom√°tica por parte de una computadora üöÄ.

## üõ†Ô∏è ¬øQu√© es un IDE?

Un **IDE** (*Integrated Development Environment* o *Entorno de Desarrollo Integrado*) es una aplicaci√≥n que proporciona a los programadores un conjunto de herramientas en un solo lugar para facilitar la creaci√≥n de software.

Combina varias funciones esenciales en una sola interfaz, lo que hace que el proceso de programaci√≥n sea m√°s **eficiente, organizado y productivo**.

---

## üß© Componentes Principales de un IDE

- **üìù Editor de C√≥digo**  
  √Årea donde se escribe el c√≥digo fuente. Incluye funciones como:
  - Resaltado de sintaxis
  - Autocompletado
  - Sugerencias inteligentes

- **üêû Depurador (Debugger)**  
  Permite ejecutar el c√≥digo paso a paso para detectar errores. Muestra el estado de las variables y el flujo del programa en tiempo real.
  
- **üóÇÔ∏è Explorador de Proyectos**  
  Organiza y permite navegar por los archivos y carpetas del proyecto de forma estructurada.

- **üî® Herramientas de Construcci√≥n (Build Tools)**  
  Automatizan tareas como la compilaci√≥n, ejecuci√≥n de pruebas y empaquetado del software.

- **üîÑ Control de Versiones**  
  Integraci√≥n con sistemas como **Git**, para rastrear cambios, colaborar y gestionar versiones del proyecto.

---

## üíª Ejemplos Populares de IDEs

- **Visual Studio Code (VSCode)** ‚Äì Ligero, extensible y muy popular.
- **Jupyter Notebook** ‚Äì Ideal para ciencia de datos y visualizaci√≥n interactiva.

## üìö ¬øQu√© es una librer√≠a?

En programaci√≥n, una **librer√≠a** es una colecci√≥n de funciones, clases y recursos predefinidos que los desarrolladores pueden utilizar para realizar tareas comunes sin tener que escribir todo el c√≥digo desde cero.

Las librer√≠as permiten **ahorrar tiempo y esfuerzo**, ya que ofrecen soluciones probadas y optimizadas para problemas frecuentes.

---

## üß© Caracter√≠sticas de una Librer√≠a

- **üîÅ Reutilizaci√≥n de C√≥digo**  
  Permiten aprovechar c√≥digo ya existente, reduciendo la duplicaci√≥n de esfuerzos y mejorando la eficiencia.

- **üì¶ Modularidad**  
  Est√°n organizadas en m√≥dulos, lo que permite importar solo las partes necesarias para un proyecto espec√≠fico.

- **üõ†Ô∏è Mantenimiento y Actualizaci√≥n**  
  Suelen ser mantenidas por comunidades o empresas, lo que garantiza mejoras constantes, seguridad y compatibilidad con nuevas versiones del lenguaje.

- **üìñ Facilidad de Uso**  
  Vienen acompa√±adas de documentaci√≥n y ejemplos, lo que facilita su integraci√≥n en proyectos nuevos o existentes.


```python
# Librerias standard
import os
import warnings
warnings.filterwarnings('ignore')

# Manipulaci√≥n de datos y an√°lisis
import pandas as pd
import numpy as np

# Visualizaci√≥n de datos
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning - Preprocesamiento
from sklearn.preprocessing import LabelEncoder, StandardScaler, PolynomialFeatures
from sklearn.impute import SimpleImputer

# Machine Learning - Modelado y evaluaci√≥n
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, f1_score

# Machine Learning - Modelos
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier

# Persistencia de modelo
import joblib

df = pd.read_csv("personality_datasert.csv")
df.head()
```

```python
# Diccionario de nombres nuevos (todo en min√∫scula y en espa√±ol)
nombres_columnas = {
    'Time_spent_Alone': 'tiempo_solo',
    'Stage_fear': 'miedo_escenico',
    'Social_event_attendance': 'asistencia_eventos_sociales',
    'Going_outside': 'salir_afuera',
    'Drained_after_socializing': 'agotado_despues_socializar',
    'Friends_circle_size': 'tamano_circulo_amigos',
    'Post_frequency': 'frecuencia_publicaciones',
    'Personality': 'personalidad'
}
# Renombrar columnas
df.rename(columns=nombres_columnas, inplace=True)
# Mapear valores a espa√±ol
valores_esp= {'Yes': 'Si', 'No': 'No'}

df['miedo_escenico'] = df['miedo_escenico'].map(valores_esp)
df['agotado_despues_socializar'] = df['agotado_despues_socializar'].map(valores_esp)

df['personalidad'] = df['personalidad'].map({
    'Extrovert': 'extrovertido',
    'Introvert': 'introvertido'
})
df.head()
```

## 3. üßπ Preprocesamiento de Datos <a name="preprocesamiento"></a>

En esta secci√≥n:
- Imputamos valores faltantes
- Codificamos variables categ√≥ricas
- Normalizamos variables num√©ricas

```python
#Como vimos anteriormente, el dataset no presenta nulos.
```

```python
# Vamos a pasar la variable target a valor num√©rico:
le = LabelEncoder()
df[target] = le.fit_transform(df[target])
print(f"Clases encodeadas: {dict(zip(le.classes_, le.transform(le.classes_)))}")
```

```python
df.personalidad.head()
```

```python
df[categorical_cols].head()
```

```python
# Encodedo features categ√≥ricas
cols_antes = set(df.columns)
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
cols_despues = set(df.columns)
categoricas = list(cols_despues - cols_antes)
print("Nuevas variables dummy creadas:")
print(categoricas)
```

```python
# Topeo los OutLiars
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1 #Rango Intercuartilico
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=limite_inf, upper=limite_sup)
```

```python
df.head()
```

```python
# Train-test split
X = df.drop(columns=[target])
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
X_train.head()
```

```python
scaler = StandardScaler()

# Escalamos solo las num√©ricas en train
X_train_num = pd.DataFrame(
    scaler.fit_transform(X_train[numeric_cols]),
    columns=numeric_cols,
    index=X_train.index
)

# Aplicamos la misma transformaci√≥n a test
X_test_num = pd.DataFrame(
    scaler.transform(X_test[numeric_cols]),
    columns=numeric_cols,
    index=X_test.index
)

# Concatenamos con las categ√≥ricas sin modificar
X_train_std = pd.concat([X_train_num, X_train[categoricas]], axis=1)
X_test_std = pd.concat([X_test_num, X_test[categoricas]], axis=1)
X_train_std.head()
```

## 4. ü§ñ Entrenamiento y Evaluaci√≥n <a name="entrenamiento-evaluacion"></a>

Entrenamos con diferentes modelos y comparamos:
- GradientBoosting
- Logistic Regression

```python
# Creamos el modelo con los hiperpar√°metros fijos
clf = GradientBoostingClassifier()
# Entrenamos
clf.fit(X_train_std, y_train)

# Predicciones de probabilidad de que sea clase=1=introvertido seg√∫n label encoder
y_train_proba = clf.predict_proba(X_train_std)[:, 1]
y_test_proba = clf.predict_proba(X_test_std)[:, 1]
```

```python
auc_train = roc_auc_score(y_train, y_train_proba)
auc_test = roc_auc_score(y_test, y_test_proba)

print(f"üß™ AUC en Train: {auc_train:.4f}")
print(f"üß™ AUC en Test:  {auc_test:.4f}")
```

```python
# Definimos los modelos y sus par√°metros
models = {
    'GradientBoosting': {
        'estimator': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators': [250, 300, 350, 400, 450],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    },
    'LogisticRegression': {
        'estimator': LogisticRegression(max_iter=1000),
        'params': {
            'C': np.logspace(-4, 4, 20),
            'solver': ['lbfgs', 'liblinear']
        }
    }
}

# Entrenamiento y b√∫squeda
results = {}
for name, cfg in models.items():
    clf = RandomizedSearchCV(
        estimator=cfg['estimator'],
        param_distributions=cfg['params'],
        n_iter=20,
        cv=3,
        scoring='roc_auc',
        n_jobs=-1,
        random_state=42
    )
    clf.fit(X_train_std, y_train)
    results[name] = clf
    print(f"‚úÖ {name} - Mejor AUC (CV): {clf.best_score_:.4f}")
    print(f"üîß {name} - Mejores par√°metros: {clf.best_params_}\n")
```

```python
display(Image(filename="cv.png", width=600))
display(Markdown("*Fuente de la imagen: Omkar Hankare(2023), [Cross Validation](https://ompramod.medium.com/cross-validation-623620ff84c2), Medium.*"))
```

```python
#ENTRENAR GRADIENT BOOSTING Y CALCULAR AUC POR ETAPA
gb_best = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3,
    n_estimators=400,
    random_state=42
)
gb_best.fit(X_train_std, y_train)

# AUC por etapa
train_auc, test_auc = [], []
for y_train_pred, y_test_pred in zip(gb_best.staged_predict_proba(X_train_std), gb_best.staged_predict_proba(X_test_std)):
    train_auc.append(roc_auc_score(y_train, y_train_pred[:, 1]))
    test_auc.append(roc_auc_score(y_test, y_test_pred[:, 1]))

best_n = np.argmax(test_auc) + 1
print(f'‚úÖ GradientBoosting - Mejor AUC en test: {max(test_auc):.4f} con n_estimators = {best_n}')

#ENTRENAR NUEVO GB CON best_n
gb_best = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3,
    n_estimators=best_n,
    random_state=42
)
gb_best.fit(X_train_std, y_train)
```

```python
# Predicciones de probabilidad
y_train_proba = gb_best.predict_proba(X_train_std)[:, 1]
y_test_proba = gb_best.predict_proba(X_test_std)[:, 1]

auc_train = roc_auc_score(y_train, y_train_proba)
auc_test = roc_auc_score(y_test, y_test_proba)

print(f" AUC en Train: {auc_train:.4f}")
print(f" AUC en Test:  {auc_test:.4f}")
```

```python
# GRAFICAR AUC vs n_estimators para GB
plt.figure(figsize=(10, 6))
plt.plot(train_auc, label='Train AUC', color='blue')
plt.plot(test_auc, label='Test AUC', color='orange')
plt.axvline(best_n, linestyle='--', color='gray', label=f'Mejor n = {best_n}')
plt.xlabel(f'N√∫mero de √°rboles {best_n}')
plt.ylabel('AUC')
plt.title('Curva de Aprendizaje - Gradient Boosting')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# AC√Å VEMOS QUE PASA SI SEGUIMOS AUMENTANDO EL N_ESTIMATORS EMPEZAMOS A PERDER GENERALIZACI√ìN
gb_overfit = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3,
    n_estimators=1000,
    random_state=42
)
gb_overfit.fit(X_train_std, y_train)

# AUC por etapa
train_auc_of, test_auc_of = [], []
for y_train_pred, y_test_pred in zip(gb_overfit.staged_predict_proba(X_train_std), gb_overfit.staged_predict_proba(X_test_std)):
    train_auc_of.append(roc_auc_score(y_train, y_train_pred[:, 1]))
    test_auc_of.append(roc_auc_score(y_test, y_test_pred[:, 1]))

#ENTRENAR NUEVO GB CON best_n
gb_overfit = GradientBoostingClassifier(
    learning_rate=0.01,
    max_depth=3,
    n_estimators=1000,
    random_state=42
)
gb_overfit.fit(X_train_std, y_train)
```

```python
# GRAFICAR AUC vs n_estimators para GB overfiteado
plt.figure(figsize=(10, 6))
plt.plot(train_auc_of, label='Train AUC', color='blue')
plt.plot(test_auc_of, label='Test AUC', color='orange')
plt.axvline(370, linestyle='--', color='gray')
plt.xlabel(f'N√∫mero de √°rboles {1000}')
plt.ylabel('AUC')
plt.title('Curva de Aprendizaje 1000n- Gradient Boosting')
plt.legend()
plt.grid(True)
plt.show()
```
